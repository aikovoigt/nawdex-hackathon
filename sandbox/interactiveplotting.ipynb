```python
from ipywidgets import interactive

import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

```python
from dask.distributed import Client
client = Client()
client
```

```python
def load_griddata(gridres):
    path  = '/scratch/b/b380459/icon_4_hackathon/'
    fname = path+'/grids/icon-grid_nawdex_78w40e23n80n_'+gridres+'.nc'
    return xr.open_dataset(fname)[['clon','clat','cell_area']].rename({'cell': 'ncells'}) 

def load_2d(expid):
    path  = '/scratch/b/b380459/icon_4_hackathon/'
    fname = path+'/'+expid+'/'+expid+'_2016*_2d_30min_DOM01_ML_0*.nc'
    return ( xr.open_mfdataset(fname,
                               combine='by_coords',parallel=True, 
                               #engine='h5netcdf', 
                               chunks={'time': 1}) )

def load_data(expid, gridres):
    return xr.merge([load_2d(expid), load_griddata(gridres)])
```

```python
ds = load_data('nawdexnwp-80km-mis-0001', 'R80000m')
ds = ds.drop(['lev_bnds', 'lev_2_bnds', 'lev_3_bnds', 'lev_2', 'lev_3', 'lev', 'height', 'height_2']).squeeze()
```

```python
ds
```

```python
def plot_field(field,time):
    var=ds[field].isel(time=time)
    plt.figure(figsize=(6,4))
    plt.tricontourf(np.rad2deg(ds.clon), np.rad2deg(ds.clat), var)
    plt.colorbar()
    plt.show()
    return

w = interactive(plot_field, field=ds.data_vars.keys(), time=(0,len(ds.time)-1,1))

output = w.children[-1]
output.layout.height = '500px'
display(w)
```
