```python
import numpy as np
import xarray as xr
import os
import matplotlib.pyplot as plt
```

```python
import dicthelpers as dh
color_res = dh.plotcolors()
simdict = dh.simdictionary()
```

```python
path = '/work/bb1018/icon_4_hackathon/ccf/'
```

```python
ds_list1 = []; ds_list2 = []
for filename in os.listdir(path):
    if filename.endswith("_omega500_pdf.nc"):
        ds_temp = xr.open_dataset(path+filename)
        ds_list1.append(ds_temp)
        ds_temp.close()
    if filename.endswith("_clct_sampled_on_omega500.nc"):
        ds_temp = xr.open_dataset(path+filename)
        ds_list2.append(ds_temp)
        ds_temp.close()        
```

```python
ds_list = []
for ind in range(len(ds_list1)):
    ds_list.append(xr.merge([ds_list1[ind], ds_list2[ind]]))

del ds_list1, ds_list2
```

```python
plt.figure(figsize=(12,4))

pan1=plt.subplot(2,2,1)
pan2=plt.subplot(2,2,2)
pan3=plt.subplot(2,2,3)
pan4=plt.subplot(2,2,4)
for ds in ds_list:
    simname = ds.omega_pdf.attrs['simulation']
    res     = (simdict[simname])['res']
    start   = (simdict[simname])['start']
    conv    = (simdict[simname])['conv' ] 
    mphys   = (simdict[simname])['mphys']
    if conv == 0: linestyle='--'
    if conv == 1: linestyle='-'
    if start == '20160922':
        if mphys == 1:
            pan1.plot(ds.bins, ds['omega_pdf'], color=color_res[res], linestyle=linestyle, label=res)
            pan2.plot(ds.bins, ds['clct_sampled'], color=color_res[res], linestyle=linestyle)
        if mphys == 2:
            pan3.plot(ds.bins, ds['omega_pdf'], color=color_res[res], linestyle=linestyle)
            pan4.plot(ds.bins, ds['clct_sampled'], color=color_res[res], linestyle=linestyle)            

pan1.legend()

plt.show()
```
